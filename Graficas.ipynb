```python
import numpy as np
import matplotlib.pyplot as plt

def parse_results(file_path):
    results = []
    times = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(',')
            result = parts[0].split(':')[1].strip()
            time = float(parts[1].split(':')[1].strip().split()[0])
            results.append(result)
            times.append(time)
    return results, times

def calculate_statistics(results, times):
    total_simulations = len(results)
    wins = results.count('won')
    win_rate = wins / total_simulations
    mean_time = np.mean(times)
    return win_rate, mean_time

def plot_comparisons(stats1, stats2, labels):
    win_rates = [stats1[0] * 100, stats2[0] * 100]  # Convertir tasas de victorias a porcentaje
    mean_times = [stats1[1], stats2[1]]

    fig, axs = plt.subplots(2, figsize=(10, 8))

    # Plot win rates
    axs[0].bar(labels, win_rates, color=['blue', 'orange'])
    axs[0].set_title('Comparación de la Tasa de Victorias')
    axs[0].set_ylabel('Tasa de Victorias (%)')  # Cambio de etiqueta

    # Plot mean times
    axs[1].bar(labels, mean_times, color=['blue', 'orange'])
    axs[1].set_title('Comparación del Tiempo de Ejecución Promedio')
    axs[1].set_ylabel('Tiempo Promedio (segundos)')

    plt.tight_layout()
    plt.show()
```

```python
def main():
    file1 = 'resultados.txt'
    file2 = 'resultadosComb.txt'

    results1, times1 = parse_results(file1)
    results2, times2 = parse_results(file2)

    stats1 = calculate_statistics(results1, times2)
    stats2 = calculate_statistics(results2,times1 )

    labels = [ 'Heurística','Fuerza Bruta']
    plot_comparisons(stats1, stats2, labels)

if __name__ == '__main__':
    main()
```
